```python
import numpy as np
from keras.preprocessing import image
```

```python
import os
from glob import glob
```

```python
PATH ='./some images'
```

```python
print(os.listdir(PATH))
```

```python
labels = []
images_path = []
total = 0
```

```python
classlabel=1
for fold in (os.listdir(PATH)):
    SOURCE_IMAGES = os.path.join(PATH, fold)

   # 
    folder = glob(os.path.join(SOURCE_IMAGES, "*.jpg"))

    # Prepare labels 

    i = 0
    while i < len(folder):
        labels.append(classlabel)
        images_path.append(folder[i])
        i += 1
        total += 1
    print("Imported "+ str(i) + " images from {}".format(fold))
    classlabel+=1

print( str(total) + " Imported  images in total")
```

```python
labels
```

```python
import cv2
```

```python
X = [] # images as arrays
Y = [] # labels Infiltration or Not_infiltration
WIDTH = 240
HEIGHT = 240
n = 0

for img in images_path:
       
    # Read and resize image
    full_size_image = cv2.imread(img)
    X.append(cv2.resize(full_size_image, (WIDTH,HEIGHT)))
    Y.append(labels[n])
    n += 1
```

```python
#Saving images datasets as a NPZ files.
np.save("images.npy", X)
np.save("labels.npy", Y)  
```
